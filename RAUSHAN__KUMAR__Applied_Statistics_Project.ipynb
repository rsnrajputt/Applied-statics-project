##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


```python
# Calculate measures of central tendency
mean_values = df[['Year_Birth', 'Income']].mean()
median_values = df[['Year_Birth', 'Income']].median()
mode_values = df[['Year_Birth', 'Income']].mode().iloc[0]

# Calculate measures of dispersion
variance_values = df[['Year_Birth', 'Income']].var()
std_dev_values = df[['Year_Birth', 'Income']].std()

# Identify outliers using z-scores
from scipy.stats import zscore

z_scores = zscore(df[['Year_Birth', 'Income']])
outliers = (abs(z_scores) > 3).any(axis=1)

# Handle outliers by replacing them with the median
df.loc[outliers, ['Year_Birth', 'Income']] = df[['Year_Birth', 'Income']].median()

# Display the calculated values
mean_values, median_values, mode_values, variance_values, std_dev_values
```

**Deliverables**:

- **Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.**:

  

```python
# Display the descriptive statistics
descriptive_statistics = pd.DataFrame({
    'Measure': ['Mean', 'Median', 'Mode', 'Variance', 'Std Deviation'],
    'Year_Birth': [mean_values['Year_Birth'], median_values['Year_Birth'], mode_values['Year_Birth'], variance_values['Year_Birth'], std_dev_values['Year_Birth']],
    'Income': [mean_values['Income'], median_values['Income'], mode_values['Income'], variance_values['Income'], std_dev_values['Income']]
})

descriptive_statistics
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



```python
import numpy as np
from scipy.stats import binom, norm

# Example variable: 'NumDealsPurchases'
variable = 'NumDealsPurchases'

# Identify distribution type
# For example, if the variable represents the number of deals, it could follow a Binomial distribution.
# If the variable is continuous and looks symmetric, it might follow a Normal distribution.

# Binomial Distribution
n = df[variable].max()  # Assuming the maximum number of deals in a purchase as the number of trials
p = df[variable].mean() / n  # Assuming the mean number of deals as the probability of success
binomial_dist = binom(n, p)

# Calculate probabilities
prob_0 = binomial_dist.pmf(0)
prob_1 = binomial_dist.pmf(1)
prob_2_or_more = 1 - binomial_dist.cdf(1)

# Calculate expected value (mean) for Binomial distribution
expected_value_binomial = binomial_dist.mean()

# Normal Distribution (for illustration purposes, assuming the variable is approximately normally distributed)
mu = df[variable].mean()
sigma = df[variable].std()
normal_dist = norm(mu, sigma)

# Calculate probabilities for a Normal distribution
prob_below_2 = normal_dist.cdf(2)
prob_above_3 = 1 - normal_dist.cdf(3)

# Calculate expected value (mean) for Normal distribution
expected_value_normal = normal_dist.mean()

# Display results
probabilities = pd.DataFrame({
    'Event': ['0 ' + variable, '1 ' + variable, '2 or more ' + variable, '< 2 ' + variable, '> 3 ' + variable],
    'Probability': [prob_0, prob_1, prob_2_or_more, prob_below_2, prob_above_3]
})

expected_values = pd.DataFrame({
    'Distribution': ['Binomial', 'Normal'],
    'Expected Value': [expected_value_binomial, expected_value_normal]
})

probabilities, expected_values
```

## Task 4: Insights and Customer Segmentation üìà

- **Explore relationships between customer characteristics and spending habits. Segment customers based on their behaviors and characteristics.**

**Deliverables**:

- **Key insights regarding relationships between variables and distinct customer segments based on behaviors.**

  

```python
# Explore relationships between customer characteristics and spending habits

# Correlation matrix to identify relationships between variables
correlation_matrix = df[['Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds']].corr()

# Heatmap for visualization
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Explore customer segmentation based on behaviors and characteristics

# For illustration purposes, let's use K-means clustering with selected features
from sklearn.cluster import KMeans

# Select features for clustering
selected_features = df[['MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds']]

# Standardize features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(selected_features)

# Determine the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()

# Based on the Elbow Method, let's choose an optimal K (number of clusters) and fit the model
optimal_k = 3
kmeans_model = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=0)
df['Cluster'] = kmeans_model.fit_predict(scaled_features)

# Display insights and customer segments
segmentation_insights = df.groupby('Cluster').mean()[['Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds']]

segmentation_insights
```
